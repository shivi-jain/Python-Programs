```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://clutch.co/agencies/digital-marketing'
response = requests.get(url)
content = BeautifulSoup(response.content, 'lxml')
```

```python
print('__________________________________Company Name______________________________________')
for Company_Name in content.findAll('h3', attrs={"class": "company-name"}):
      print(Company_Name.text.encode('utf-8'))

print('____________________List Items(Min. Project size, hrly rate, Employees and location_____________________')
for Content in content.findAll('div', attrs={"class": "list-item"}):
    print(Content.text.encode('utf-8')) 

print('__________________________________Ratings______________________________________')
for Span in content.findAll('span', attrs={"class": "rating"}):
        print(Span.text.encode('utf-8'))
        
print('__________________________________Reviews______________________________________')
for Link in content.findAll('a', attrs={"target": "_blank"}):
    a = []
    line = Link.text.encode('utf-8')
    for word in line.split():
        try:
            a.append(int(word))
            print(a)
        except ValueError:
            pass
        
print('__________________________________All links______________________________________')
for Link in content.findAll('a', attrs={"target": "_blank"}):
        print(Link.get("href"))
        
print('__________________________________Description______________________________________')
for Description in content.findAll('p'):
        print(Description.text.encode('utf-8'))
```
